```python
import json
import time

fd = open('records.json','r')
js = fd.read()
fd.close()


record = json.loads(js)

print("---------------MENU---------------\n")

for key in record:
    print(key,':',record[key]['Name'],',',record[key]['Price'],',',record[key]['Qn'])
    
print("\n----------------------------------\n")

user_nm = input("Enter your name                :  ")
user_cn = input("Enter your contact number      :  ")
prod_id = input("Enter the product ID           :  ")
prod_qn = int(input("Enter the product Quantity :  "))

if(record[prod_id]['Qn'] >= prod_qn):
    
    record[prod_id]['Qn'] = record[prod_id]['Qn'] - prod_qn

    print("\n----------------------------------")
    print("              Bill              \n")
    print("Product Name     :  ",record[prod_id]['Name'])
    print("Product Price    :  ",record[prod_id]['Price'])
    print("Product Quantity :  ",prod_qn)
    print("\n----------------------------------")
    billing_amt = prod_qn*int(record[prod_id]['Price'])
    if(billing_amt >= 1000):
        print("Yay! You have got a 10% discount on your total purchase!")
        billing_amt = billing_amt - int(billing_amt * 0.1)
        print("Billing Amount :  ",billing_amt)
    else:
        print("Billing Amount :  ",billing_amt)
    print("----------------------------------")
    
    fd = open("sales.txt",'a')
    user_info = user_nm + ', ' + user_cn + ',' + prod_id + ',' +  record[prod_id]['Name'] + ',' +  str(prod_qn) + ',' +  str(billing_amt) + ',' +  time.ctime() + '\n'
    fd.write(user_info)
    fd.close()
    
    
else:
    
    if(record[prod_id]['Qn'] != 0):
        print("Sorry we do not have enough quantity of the product.")
        print("Available quantity : ",record[prod_id]['Qn'])
        print("Would you like to purchase it?")
        
        ch = input(r"Y/N : ")
        
        if(ch=='Y' or ch=='y'):
            print("\n----------------------------------")
            print("              Bill              \n")
            print("Product Name     :  ",record[prod_id]['Name'])
            print("Product Price    :  ",record[prod_id]['Price'])
            print("Product Quantity :  ",record[prod_id]['Qn'])
            print("\n----------------------------------")
            billing_amt = prod_qn*int(record[prod_id]['Price'])
            if(billing_amt >= 1000):
                print("Yay! You have got a 10% discount on your total purchase!")
                billing_amt = billing_amt - int(billing_amt * 0.1)
                print("Billing Amount :  ",billing_amt)
            else:
                print("Billing Amount :  ",billing_amt)
            print("----------------------------------")
            
            fd = open("sales.txt",'a')
            user_info = user_nm + ', ' + user_cn + ',' + prod_id + ',' +  record[prod_id]['Name'] + ',' +  str(record[prod_id]['Qn']) + ',' +  str(record[prod_id]['Qn']*int(record[prod_id]['Price'])) + ',' +  time.ctime() + '\n'
            fd.write(user_info)
            fd.close()
            
            record[prod_id]['Qn'] = 0
            
        else:
            print("Thanks")
            
    else:
        print("Sorry, the required product is not currently available.")
    

js = json.dumps(record)

fd = open('records.json','w')
fd.write(js)
fd.close()
```
