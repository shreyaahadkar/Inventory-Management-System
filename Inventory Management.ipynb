```python
import time

#Reading the inventory
fd = open('inventory.txt','r')
products = fd.read().split('\n')
fd.close()

#Taking user input
user_name = input("Enter your name            :  ")
user_phone= input("Enter your phone no.       :  ")
user_mail = input("Enter your mail ID         :  ")
prod_id   = input("Enter the product ID       :  ")
prod_qn   = input("Enter the product quantity :  ")

updated_prod = []

#Going through each produt detail

for product in products:
    
    product = product.split(',')
    
    if(product[0]==prod_id):
        
        #Checking if the product exists
        
        if(int(product[3]) >= int(prod_qn)):
            
            #If we have enough quantity
            
            print("---------------------")
            print("Product Name   :  ",product[1])
            print("Product Price  :  ",product[2])
            print("Quantity       :  ",prod_qn)
            print("---------------------")
            billing_amt = int(product[2])*int(prod_qn)
            if(billing_amt >= 1000):
                print("Yay! You have got a 10% discount on your total purchase!")
                billing_amt = billing_amt - int(billing_amt * 0.1)
                print("Billing Amount :  ",billing_amt)
            else:
                print("Billing Amount :  ",billing_amt)
            print("---------------------")
            
            product[3] = str(int(product[3]) - int(prod_qn))
            
            #Creating a sales record file
            
            fd = open("sales.txt",'a')
            user_sales = user_name+', '+user_phone+', '+user_mail+', '+product[1]+', '+prod_qn+', '+str(int(product[2])*int(prod_qn))+', '+time.ctime()+"\n"
            fd.write(user_sales)
            fd.close()
            
        else:
            
            #If not
            
            print("Sorry we do not have enough quantity")
            print("Available quantity:  ",product[3])
            print('Would you like to purchase it?')
            
            ch = input("Y\\N: ")
            
            if(ch=='Y' or ch=='y'):
                
                print("---------------------")
                print("Product Name   :  ",product[1])
                print("Product Price  :  ",product[2])
                print("Quantity       :  ",product[3])
                print("---------------------")
                billing_amt = int(product[2])*int(prod_qn)
                if(billing_amt >= 1000):
                    print("Yay! You have got a 10% discount on your total purchase!")
                    billing_amt = billing_amt - int(billing_amt * 0.1)
                    print("Billing Amount :  ",billing_amt)
                else:
                    print("Billing Amount :  ",billing_amt)
                print("---------------------")
                
                fd = open("sales.txt",'a')
                user_sales = user_name+', '+user_phone+', '+user_mail+', '+product[1]+', '+product[3]+', '+str(int(product[2])*int(product[3]))+', '+time.ctime()+"\n"
                fd.write(user_sales)
                fd.close()
        
                product[3] = '0'
            else:
                print("Thanks")
            
    
    updated_prod.append(product)

#Updating the inventory    

fd = open('inventory.txt','w')

for i in updated_prod[:-1]:
    prod = i[0]+','+i[1]+','+i[2]+','+i[3]+'\n'
    fd.write(prod)
    
fd.close()            
print("---------------------")
print("Inventory Updated")
```
